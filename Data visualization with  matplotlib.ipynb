```python
import matplotlib.pyplot as plt
import pandas as pd 
# Sample data
data = {
    'Course': ['C', 'C++', 'JAVA', 'PYTHON', 'R'],
    'Students enrolled': [40, 15, 30, 50, 20]
}
 
# Create a DataFrame from the data
df = pd.DataFrame(data)
 
# Create a bar chart using Matplotlib
plt.bar(df['Course'],df['Students enrolled'],width=0.5, color="green")
#title
plt.title('Students intrest in programming lang')
#labels
plt.xlabel('Course name')
plt.ylabel('No.Of Students')
#Legend
plt.legend('S')
# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd 
# Sample data
data = {
    'Course': ['C', 'C++', 'JAVA', 'PYTHON', 'R'],
    'Students enrolled': [40, 15, 30, 50, 20]
}
 
# Create a DataFrame from the data
df = pd.DataFrame(data)
 
# Create a line chart using Matplotlib
plt.plot(df['Course'],df['Students enrolled'], color="maroon")
#title
plt.title('Students intrest in programming lang')
#labels
plt.xlabel('Course name')
plt.ylabel('No.Of Students')
#Legend
plt.legend('S')
# Display the chart
plt.show()
```
